```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
img_path = 'train/picture/LFPW_image_train_0871_3.jpg'
a = cv2.imread(img_path)
landmarks = []
with open('train/landmark/LFPW_image_train_0871_3.txt') as f:
    lines = f.readlines()
for line in lines[1:]:
    x, y = line.strip().split(' ')
    x, y = int(float(x)), int(float(y))
    landmarks.append([x, y])
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
for i, lm in enumerate(landmarks):
    x, y = lm
    a = cv2.circle(a, (x, y), 1, (0, 255, 0), 4)
#     a = cv2.putText(a, str(i), (x, y), font, 1, (0, 255, 0), 2)
```

```python
cv2.imwrite('test.jpg', a)
```
